```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(42)
```

## Load Dataset

```python
# Create synthetic binary classification dataset
m_train = 200
m_test = 50
n_features = 20

# Generate data with some structure (not purely random)
np.random.seed(42)

# Training data
X_train = np.random.randn(n_features, m_train)
true_weights = np.random.randn(n_features, 1) * 0.5
y_train_logits = np.dot(true_weights.T, X_train) + 0.5
y_train = (y_train_logits > 0).astype(int)

# Test data
X_test = np.random.randn(n_features, m_test)
y_test_logits = np.dot(true_weights.T, X_test) + 0.5
y_test = (y_test_logits > 0).astype(int)

print(f"Training set: X_train.shape = {X_train.shape}, y_train.shape = {y_train.shape}")
print(f"Test set: X_test.shape = {X_test.shape}, y_test.shape = {y_test.shape}")
print(f"Training labels - 0s: {np.sum(y_train == 0)}, 1s: {np.sum(y_train == 1)}")
```

## Helper Functions

```python
def sigmoid(z):
    """
    Compute sigmoid activation
    """
    # üêõ BUG #1: Numerical instability with large values
    # This can cause overflow/underflow issues
    z = np.clip(z, -500, 500)  # prevent overflow
    return 1 / (1 + np.exp(-z))
```

```python
def initialize_parameters(n_features):
    """
    Initialize weights and bias
    """
    # üêõ BUG #2: Bad initialization - weights too large!
    # This causes gradients to explode
    w = np.random.randn(n_features, 1) * 0.01
    b = 0.0
    return w, b
```

```python
def propagate(w, b, X, y):
    """
    Forward and backward propagation
    """
    m = X.shape[1]

    # Forward propagation
    A = sigmoid(np.dot(w.T, X) + b)

    # üêõ BUG #3: Wrong cost calculation!
    # Missing the negative sign in cross-entropy
    # Should be: cost = -1/m * np.sum(y * np.log(A) + (1 - y) * np.log(1 - A))
    epsilon = 1e-15
    A = np.clip(A, epsilon, 1 - epsilon)
    cost = -1/m * np.sum(y * np.log(A) + (1 - y) * np.log(1 - A))

    # üêõ BUG #4: Wrong gradient calculation!
    # Signs are flipped - should be (A - y), not (y - A)
    dw = 1/m * np.dot(X, (A - y).T)
    db = 1/m * np.sum(A - y)

    grads = {"dw": dw, "db": db}
    return grads, cost
```

## Training Function

```python
def train(w, b, X, y, num_iterations, learning_rate, print_cost=True):
    """
    Optimize parameters using gradient descent
    """
    costs = []

    for i in range(num_iterations):
        # Get gradients and cost
        grads, cost = propagate(w, b, X, y)

        dw = grads["dw"]
        db = grads["db"]

        # Update parameters
        w = w - learning_rate * dw
        b = b - learning_rate * db

        # Record cost
        if i % 100 == 0:
            costs.append(cost)
            if print_cost:
                print(f"Cost after iteration {i}: {cost:.6f}")

    params = {"w": w, "b": b}
    return params, costs
```

```python
def predict(w, b, X):
    """
    Predict labels for dataset X
    """
    m = X.shape[1]
    y_pred = np.zeros((1, m))

    A = sigmoid(np.dot(w.T, X) + b)

    for i in range(m):
        if A[0, i] > 0.5:
            y_pred[0, i] = 1
        else:
            y_pred[0, i] = 0

    return y_pred
```

## Run the Model

Watch what happens when you run this... üî•

```python
# Initialize parameters
w, b = initialize_parameters(n_features)

print(f"Initial w stats: mean={np.mean(w):.3f}, std={np.std(w):.3f}, max={np.max(np.abs(w)):.3f}")
print(f"Initial b: {b}")
print()

# Set learning rate
learning_rate = 0.01
num_iterations = 2000

# Train the model
print("Training the model...\n")
try:
    params, costs = train(w, b, X_train, y_train, num_iterations, learning_rate)
    print("\n‚úÖ Training completed!")
except Exception as e:
    print(f"\n‚ùå Training failed with error: {e}")
    print("\nThis is expected with these bugs. Try to fix them!")
    params = {"w": w, "b": b}
    costs = []
```

```python
# Tool 3: Numerical gradient checking
def check_gradient(w, b, X, y, epsilon=1e-7):
    """
    Check if analytical gradient matches numerical gradient
    """
    # Get analytical gradient
    grads, _ = propagate(w, b, X, y)

    # Compute numerical gradient for first weight
    w_plus = w.copy()
    w_plus[0, 0] += epsilon
    _, cost_plus = propagate(w_plus, b, X, y)

    w_minus = w.copy()
    w_minus[0, 0] -= epsilon
    _, cost_minus = propagate(w_minus, b, X, y)

    numerical_grad = (cost_plus - cost_minus) / (2 * epsilon)
    analytical_grad = grads["dw"][0, 0]

    print(f"\nGradient Check:")
    print(f"Numerical gradient: {numerical_grad:.8f}")
    print(f"Analytical gradient: {analytical_grad:.8f}")
    print(f"Difference: {abs(numerical_grad - analytical_grad):.8f}")

    if abs(numerical_grad - analytical_grad) < 1e-5:
        print("‚úÖ Gradients match!")
    else:
        print("‚ùå Gradients don't match - bug in gradient calculation!")

# Example usage:
# w_test, b_test = initialize_parameters(n_features)
# check_gradient(w_test, b_test, X_train[:, :10], y_train[:, :10])
```

## üìù BUG REPORT

Documenting all identified bugs and their fixes.

---

### Bug #1:
**Location:** sigmoid() function

**What's wrong:** The sigmoid function is implemented as:  
                      return 1 / (1 + np.exp(-z))

**Why it's wrong:**  

For large negative z, np.exp(-z) overflows.  
For large positive z, np.exp(-z) underflows.  

This leads to: Overflow warnings, NaN values, Training instability  

**How to fix:**  
def sigmoid(z):  
    z = np.clip(z, -500, 500)  
    return 1 / (1 + np.exp(-z))  

**Observable symptoms:** Runtime overflow warnings, Cost becomes NaN, Model fails to converge  

---

### Bug #2:
**Location:** initialize_parameters()

**What's wrong:** The weights are initialized as:  
                      w = np.random.randn(n_features, 1) * 10  

Weights are initialized with very large magnitude.

**Why it's wrong:**  

Large initial weights push the sigmoid activation into saturation (values very close to 0 or 1).  

When sigmoid saturates:  
- Gradients become extremely small (vanishing gradients)  
- Parameter updates become unstable  
- Learning slows down or may completely fail  

This leads to: Unstable training, Poor convergence, Slow or no learning  

**How to fix:**  
def initialize_parameters(n_features):  
    w = np.random.randn(n_features, 1) * 0.01  
    b = 0.0  
    return w, b  

**Observable symptoms:** Cost fluctuates wildly, Training unstable, Accuracy does not improve despite many iterations  

---

### Bug #3:
**Location:** propagate() ‚Äî Cost computation

**What's wrong:** The cost was implemented without the negative sign:  
                      cost = 1/m * np.sum(y * np.log(A) + (1 - y) * np.log(1 - A))

**Why it's wrong:**  

Binary cross-entropy loss must include a negative sign:

J = -1/m * Œ£ [ y log(A) + (1 - y) log(1 - A) ]

Without the negative sign:  
- The algorithm maximizes the loss instead of minimizing it  
- Gradient descent moves in the wrong direction  

This leads to: Increasing cost, Diverging training  

**How to fix:**  
epsilon = 1e-15  
A = np.clip(A, epsilon, 1 - epsilon)  

cost = -1/m * np.sum(  
    y * np.log(A) + (1 - y) * np.log(1 - A)  
)  

**Observable symptoms:** Cost increases instead of decreasing, Model fails to converge  

---

### Bug #4:
**Location:** propagate() ‚Äî Gradient computation

**What's wrong:** Gradients were computed using:  
                      dw = 1/m * np.dot(X, (y - A).T)  
                      db = 1/m * np.sum(y - A)

**Why it's wrong:**  

The correct derivative of logistic loss is:  
dw = 1/m * X (A - y).T  

Using (y - A) reverses the gradient direction.  
This makes gradient descent behave like gradient ascent.

This leads to: Increasing cost, Gradient check failure, No learning  

**How to fix:**  
dw = 1/m * np.dot(X, (A - y).T)  
db = 1/m * np.sum(A - y)  

**Observable symptoms:** Cost rises, Training diverges, Model performance does not improve  

---

### Bug #5:
**Location:** predict()

**What's wrong:** Prediction logic used incorrect threshold comparison (<= instead of >) and inverted class assignment.

**Why it's wrong:**  

Logistic regression predicts class 1 when:  
P(y = 1 | x) > 0.5  

Using <= 0.5 assigns labels incorrectly and effectively inverts predictions.

This leads to: Low accuracy even if training is correct  

**How to fix:**  
def predict(w, b, X):  
    A = sigmoid(np.dot(w.T, X) + b)  
    y_pred = (A > 0.5).astype(int)  
    return y_pred  

**Observable symptoms:** Very low accuracy, Predictions appear inverted, All samples classified into wrong class
